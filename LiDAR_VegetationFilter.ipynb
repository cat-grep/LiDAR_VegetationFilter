2. Convert LAZ to LAS (Using PDAL)

```python
import pdal, json
from pathlib import Path

las_dir = Path("lidar_raw_las")
las_dir.mkdir(exist_ok=True)

# be robust to .laz / .LAZ
laz_files = [*Path("lidar_raw_laz").glob("*.laz"), *Path("lidar_raw_laz").glob("*.LAZ")]

for laz_file in laz_files:
    base = laz_file.stem
    las_file = las_dir / f"{base}.las"

    pipe_def = {
        "pipeline": [
            {"type": "readers.las", "filename": str(laz_file).replace("\\","/")},
            {"type": "writers.las", "filename": str(las_file).replace("\\","/")}
        ]
    }

    print(f"Converting {laz_file} → {las_file}")
    try:
        pipe = pdal.Pipeline(json.dumps(pipe_def))
        # NOTE: some PDAL builds don’t expose .validate(); just execute.
        pipe.execute()
    except Exception as e:
        # Show PDAL's internal log if available—helps diagnose LAZ support issues.
        log = ""
        try:
            log = pipe.log
        except Exception:
            pass
        print(f"❌ Error converting {laz_file}: {e}")
        if log:
            print("---- PDAL LOG ----")
            print(log)
        continue
```

3. Read CRS, Reproject, Clip LAS Files to City Limits

```python
import geopandas as gpd
from shapely.geometry import shape
from shapely.ops import unary_union
from pyproj import CRS
from pathlib import Path
import json, glob, pdal

# Read the city limits GeoJSON and dissolve into a single geometry
city_limits = gpd.read_file("City_Limits_-_Round_Rock.geojson")[["geometry"]].dissolve()
if city_limits.crs is None:
    raise ValueError("❌ CRS is undefined in the city limits GeoJSON. Set it explicitly with .set_crs().")

las_dir = Path("lidar_raw_las")
las_files = glob.glob(str(las_dir / "*.las")) + glob.glob(str(las_dir / "*.LAS"))
if not las_files:
    raise RuntimeError("No LAS files found. Did the LAZ → LAS conversion succeed?")

# Read spatial reference from the first LAS file
sample_las = las_files[0]
info_pipe = pdal.Pipeline(json.dumps({
    "pipeline": [{"type": "readers.las", "filename": sample_las}]
}))
info_pipe.execute()

# Handle metadata: may be str or dict depending on PDAL version
raw_meta = info_pipe.metadata
if isinstance(raw_meta, str):
    meta = json.loads(raw_meta)
elif isinstance(raw_meta, dict):
    meta = raw_meta
else:
    raise TypeError(f"Unexpected metadata type: {type(raw_meta)}")

# Try to extract WKT from metadata (structure differs by version)
try:
    las_srs_wkt = meta["metadata"]["readers.las"]["srs"]["wkt"]
except (KeyError, TypeError):
    # Some PDAL builds return readers.las as a list
    readers = meta.get("metadata", {}).get("readers.las")
    if isinstance(readers, list) and readers:
        las_srs_wkt = readers[0]["srs"]["wkt"]
    else:
        raise RuntimeError(f"Cannot find SRS WKT in PDAL metadata for {sample_las}")

las_crs = CRS.from_wkt(las_srs_wkt)
print("CRS:", las_crs)
print("EPSG:", las_crs.to_epsg())

# Reproject city limits if CRS does not match LAS
if CRS.from_user_input(city_limits.crs) != las_crs:
    print(f"🔄 Reprojecting city limits from {city_limits.crs} to {las_crs}")
    city_limits = city_limits.to_crs(las_crs)

# Convert dissolved city limits geometry into WKT (more stable for PDAL)
geom = unary_union(city_limits.geometry)
polygon_wkt = geom.wkt

clip_dir = Path("lidar_clipped")
clip_dir.mkdir(exist_ok=True)

# Run PDAL crop pipeline for each LAS file
for i, las in enumerate(las_files, 1):
    out_path = clip_dir / Path(las).name
    print(f"🔹 Processing {i}/{len(las_files)}: {Path(las).name}")
    pipe_def = {
        "pipeline": [
            {"type": "readers.las", "filename": str(las).replace("\\", "/")},
            {"type": "filters.crop", "polygon": polygon_wkt},
            {"type": "writers.las", "filename": str(out_path).replace("\\", "/")}
        ]
    }
    try:
        pipe = pdal.Pipeline(json.dumps(pipe_def))
        # Note: some PDAL versions don’t expose .validate(); just execute directly
        pipe.execute()
        print(f"✅ Clipped: {out_path}")
    except RuntimeError as e:
        print(f"❌ Failed to process {las}: {e}")
```

4. Filter Out Tree Canopy (Class 5: High Vegetation Cloud Points)

```python
import os, glob, json
import pdal
from pathlib import Path

input_folder = clip_dir
output_folder = Path("lidar_high_vegetation")
output_folder.mkdir(exist_ok=True)
min_file_size_mb = 1

las_files = glob.glob(str(input_folder / "*.las")) + glob.glob(str(input_folder / "*.LAS"))

for las_file in las_files:
    size_mb = os.path.getsize(las_file) / (1024 * 1024)
    if size_mb < min_file_size_mb:
        print(f"⚠️ Skipping (too small): {las_file} ({size_mb:.2f} MB)")
        continue

    filename = Path(las_file).stem
    output_las = output_folder / f"{filename}_vegetation.las"
    print(f"🌿 Extracting vegetation from: {las_file}")

    pipeline_def = {
        "pipeline": [
            {"type": "readers.las", "filename": str(las_file)},
            {"type": "filters.range", "limits": "Classification[5:5]"},
            {"type": "writers.las", "filename": str(output_las)}
        ]
    }

    try:
        pipe = pdal.Pipeline(json.dumps(pipeline_def))
        pipe.execute()
        print(f"✅ Vegetation LAS saved to: {output_las}")
    except RuntimeError as e:
        print(f"❌ Failed to process {las_file}: {e}")
```
